```python
#y = 3x1 + 2x2 + noise
import numpy as np
np.random.seed(42)
```

```python
no_of_samples = 1000
X = np.random.rand(no_of_samples,2)
X
```

```python
true_weights = np.array([3,2])
noise = np.random.rand(1)
true_weights,noise
```

```python
yhat = np.dot(X,true_weights.T) + noise
yhat
```

```python
epochs = 1000
alpha = 0.01
weights = np.random.randn(2)
bias = np.random.randn(1)
weights,bias
```

```python
for _ in range(epochs):
    ypred = np.dot(X,weights) + bias
    error = ypred - yhat

    dw = (2/no_of_samples) * np.dot(X.T,error)
    dbias = (2/no_of_samples) * np.sum(error)

    weights-=alpha*dw
    bias-=alpha*dbias

    if(_%100==0):
        print("Loss : ", np.mean(np.abs(error)))
```
